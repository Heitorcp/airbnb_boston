## Correlogram

A correlogram is a type of plot which shows us any sort of relationship between two variables. These relationships are pretty easy, they evaluate if there is any linear dependence between these variables.
So a question that we might ask is: as the number of bedrooms increase so does the room price?
If so, then there is going to be some dependence between these two variables and so some correlation.
Let's check it for all numerical variables in our dataset.

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df_raw.corr(), xticklabels=df_raw.corr().columns, yticklabels=df_raw.corr().columns, cmap='RdYlGn', center=0, annot=False)

plt.show()
```

The price seems to have a higher correlation with the the variables "accommodates", "bedrooms" and "beds", but none these values seems to be high enough to be considered (in this case I am considering high correlations as 0.8 and above)

```python
from matplotlib.patches import Rectangle

# Colours for different percentiles
perc_25_colour = 'gold'
perc_50_colour = 'mediumaquamarine'
perc_75_colour = 'deepskyblue'
perc_95_colour = 'peachpuff'


fig, ax = plt.subplots(figsize=(8,8))


counts, bins, patches = ax.hist(df_raw.number_of_reviews[df_raw.number_of_reviews != 0], facecolor=perc_50_colour, edgecolor='gray')

# Change the colors of bars at the edges

#defining the percentiles
twentyfifth, seventyfifth, ninetyfifth = np.percentile(df_raw.number_of_reviews[df_raw.number_of_reviews != 0], [25, 75, 95])
for patch, leftside, rightside in zip(patches, bins[:-1], bins[1:]):
    if rightside < twentyfifth:
        patch.set_facecolor(perc_25_colour)
    elif leftside > ninetyfifth:
        patch.set_facecolor(perc_95_colour)
    elif leftside > seventyfifth:
        patch.set_facecolor(perc_75_colour)


#Set ticks at the edges of the bins
ax.set_xticks(bins.round(0))
plt.xticks(rotation=70)

# Set the graph title and axes titles
plt.title('Distribution of the Number of Reviews', fontsize=20)
plt.ylabel('Count', fontsize=15)
plt.xlabel('Number of Reviews', fontsize=15)


# Change the colors of bars at the edges
twentyfifth, seventyfifth, ninetyfifth = np.percentile(df_raw.number_of_reviews[df_raw.number_of_reviews != 0], [25, 75, 95])
for patch, leftside, rightside in zip(patches, bins[:-1], bins[1:]):
    if rightside < twentyfifth:
        patch.set_facecolor(perc_25_colour)
    elif leftside > ninetyfifth:
        patch.set_facecolor(perc_95_colour)
    elif leftside > seventyfifth:
        patch.set_facecolor(perc_75_colour)

# Calculate bar centre to display the count of data points and %
bin_x_centers = 0.5 * np.diff(bins) + bins[:-1]
bin_y_centers = ax.get_yticks()[1] * 0.25

# Display the the count of data points and % for each bar in histogram
for i in range(len(bins)-1):
    bin_label = "{0:,}".format(counts[i]) + "  ({0:,.2f}%)".format((counts[i]/counts.sum())*100)
    plt.text(bin_x_centers[i], bin_y_centers, bin_label, rotation=90, rotation_mode='anchor')

# Annotation for bar values
ax.annotate('Each bar shows count and percentage of total',
            xy=(.85,.30), xycoords='figure fraction',
            horizontalalignment='center', verticalalignment='bottom',
            fontsize=10, bbox=dict(boxstyle="round", fc="white"),
            rotation=-90)

#create legend
handles = [Rectangle((0,0),1,1,color=c,ec="k") for c in [perc_25_colour, perc_50_colour, perc_75_colour, perc_95_colour]]
labels= ["0-25 Percentile","25-50 Percentile", "50-75 Percentile", ">95 Percentile"]
plt.legend(handles, labels, bbox_to_anchor=(0.5, 0., 0.80, 0.99))

# Display the graph
plt.show()
```

## Analysing Super Hosts' Rooms Prices

```python
# defyning a function to create the csv files from the data

def create_csv(index, df, str):

    urls = []
    prices = []
    ratings = []
    values = []
    accommodates = []
    num_reviews = []

    for i in index:
        print(f"Room Price: U${df.loc[i].price}\nRoom URL: {df.loc[i].listing_url}\nOverall Rating: {df.loc[i].review_scores_rating}\nAccommodates: {df.loc[i].accommodates}\nOverall Value: {df.loc[i].review_scores_value}\nNumber of Reviews: {df.loc[i].number_of_reviews}\n")
        #appending these values to a list
        prices.append(df.loc[i].price)
        urls.append(df.loc[i].listing_url)
        ratings.append(df.loc[i].review_scores_rating)
        accommodates.append(df.loc[i].accommodates)
        values.append(df.loc[i].review_scores_value)
        num_reviews.append(df.loc[i].number_of_reviews)

    data = [prices, accommodates, ratings, values, num_reviews, urls]
    columns = ["Price", "Accommodates", "Overall Rating", "Value", "Reviews", "url"]
    final_df = pd.DataFrame.from_records(data).transpose()
    final_df.columns = columns
    final_df.set_index("url", inplace=True)
    
    #exporting the dataframe to a csv file
    final_df.to_csv(str, sep=",")
       
```

```python
import json
#defining a function that returns all amenities in the dataset
def unique_amenities():

    unique = []
    for l in df_raw.amenities:
        #converting the string representation of a list to a list
        l = json.loads(l)
        # print(l)
        for amenitie in l:
            if amenitie not in unique:
                unique.append(amenitie)
    
    return unique

len(unique_amenities())
```
